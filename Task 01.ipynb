```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_5795797[1].csv',skiprows=4)
```

```python
country_names = data['Country Name']
years = data.loc[:, '1960':'2020']  
population = years.sum(axis=1)
plt.figure(figsize=(80,160))
plt.bar(country_names, population, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Distribution by Country')
plt.xticks(rotation=90) 
plt.tight_layout()
plt.show()
```
